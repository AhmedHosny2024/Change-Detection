```python
import os
import cv2
import numpy as np
from sklearn.linear_model import LogisticRegression
# load svm model
# from sklearn.svm import SVC
# load the model random forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Function to read images from two folders and concatenate them
def load_concatenated_images(image_folder1, image_folder2):
    concatenated_images = []
    for filename1, filename2 in zip(os.listdir(image_folder1), os.listdir(image_folder2)):
        img1 = cv2.imread(os.path.join(image_folder1, filename1))
        img2 = cv2.imread(os.path.join(image_folder2, filename2))
        if img1 is not None and img2 is not None:
            # Concatenate images along the channels axis
            concatenated_image = np.concatenate((img1, img2), axis=-1)
            concatenated_images.append(concatenated_image)
    return np.array(concatenated_images)

# Function to load labels from a folder
def load_labels_from_folder(folder):
    labels = []
    for filename in os.listdir(folder):
        label = cv2.imread(os.path.join(folder, filename), cv2.IMREAD_GRAYSCALE)
        if label is not None:
            labels.append(label)
    return labels

image_folder1_train = "E:/College/Fourth Year/second term/RSSI/Project/Change-Detection/trainval/train_80_removing/A"
image_folder2_train = "E:/College/Fourth Year/second term/RSSI/Project/Change-Detection/trainval/train_80_removing/B"
labels_folder_train = "E:/College/Fourth Year/second term/RSSI/Project/Change-Detection/trainval/train_80_removing/label"

# Load concatenated images and labels
concatenated_images = load_concatenated_images(image_folder1_train, image_folder2_train)
labels = load_labels_from_folder(labels_folder_train)

# Convert images and labels to numpy arrays
concatenated_images = np.array(concatenated_images)
# falten the images
labels = np.array(labels)
# Flatten images
concatenated_images = concatenated_images.reshape(-1,6)
labels = labels.reshape(-1)

print(concatenated_images.shape)
print(labels.shape)
# Create a logistic regression model
# model = LogisticRegression()
model = RandomForestClassifier(n_estimators=10, random_state=0)
# Train the model
model.fit(concatenated_images, labels)
```

```python
from sklearn.metrics import jaccard_score
image_folder1_test = "E:/College/Fourth Year/second term/RSSI/Project/Change-Detection/trainval/test/A"
image_folder2_test = "E:/College/Fourth Year/second term/RSSI/Project/Change-Detection/trainval/test/B"
labels_folder_test = "E:/College/Fourth Year/second term/RSSI/Project/Change-Detection/trainval/test/label"
out_folder = "E:/College/Fourth Year/second term/RSSI/Project/Change-Detection/trainval/out_classifier"
jaccard_indices = []
# Predict on the test set
for filename1, filename2 ,label_filename in zip(os.listdir(image_folder1_test), os.listdir(image_folder2_test), os.listdir(labels_folder_test)):
        img1 = cv2.imread(os.path.join(image_folder1_test, filename1))
        img2 = cv2.imread(os.path.join(image_folder2_test, filename2))
        label = cv2.imread(os.path.join(labels_folder_test, label_filename), cv2.IMREAD_GRAYSCALE)
        if img1 is not None and img2 is not None:
            # Concatenate images along the channels axis
            concatenated_image = np.concatenate((img1, img2), axis=-1)
            concatenated_image = np.array(concatenated_image)
            # falten the images
            label = np.array(label)
            # Flatten images
            concatenated_image = concatenated_image.reshape(-1,6)
            label = label.reshape(-1)
            img_pred = model.predict(concatenated_image)
            jaccard_index = jaccard_score(label, img_pred,pos_label=255,zero_division=1)
            img_pred = img_pred.reshape(-1,256,256)
            y_test = label.reshape(-1,256,256)
            # write images
            img_pred=img_pred.squeeze()
            y_test=y_test.squeeze()
            img_pred = cv2.imwrite(os.path.join(out_folder, label_filename), img_pred)
            # # show images
            # cv2.imshow("image", img_pred)
            # cv2.imshow("label", y_test)
            # cv2.waitKey(0)
            # cv2.destroyAllWindows()
            jaccard_indices.append(jaccard_index)
            print(jaccard_index)
mean_jaccard_index = np.mean(jaccard_indices)            
print(mean_jaccard_index)
```
